```python
def Circle(ax, x0, y0, R, color='cyan'):
    
    theta = np.linspace(0, 2*np.pi, 1000)
    
    x = x0 + R*np.cos(theta)
    y = y0 + R*np.sin(theta)
    
    ax.plot(x, y, color=color)
    
    return
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(img)
ax.axis('equal')
Circle(ax, 1490, 2130, 660)
Circle(ax, 1450, 2185, 800)
Circle(ax, 1450, 2185, 980)
Circle(ax, 1450, 2130, 1250)
ax.grid()
#ax.axvline(1500)
```
